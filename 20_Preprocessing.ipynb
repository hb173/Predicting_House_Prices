```python
import pandas as pd
import numpy as np
from pathlib import Path
import zipfile


#importing the california data
PARENT = "Predicting_House_Prices"
path = Path(PARENT).parent / "../Data/cleaned_data.csv"
data = pd.read_csv(path)
data
```

```python
#removing the Unamed:0 column and the price
data2 = data.drop(columns=
    [
        "Unnamed: 0"
    ]
)
data2.sample(15)
```

```python
#removing price outliers
data2 = data2[data2["price"] > 10000]
data2
```

```python
#making all city names lower case
data2["city"]= data2["city"].str.lower()
data2
```

# Preprocessing
I will be standardizing the folowing numeric variables: livingArea, bathrooms, bedrooms, buildingArea, and garageSpaces.
While I will be one-hot encoding the categorical variables: state, county, city, event, levels, homeType, yearBuilt, and lotAreaUnits.

```python
#state ohe
data3 = pd.concat([data2,pd.get_dummies(data2['state'], prefix='state')],axis=1)
data3
```

```python
#city ohe
data4 = pd.concat([data3,pd.get_dummies(data2['city'], prefix='city')],axis=1)
data4
```

```python
#event ohe
data5 = pd.concat([data4,pd.get_dummies(data2['event'], prefix='event')],axis=1)
data5
```

```python
#lotAreaUnits ohe
data6 = pd.concat([data5,pd.get_dummies(data2['lotAreaUnits'], prefix='lotAreaUnits')],axis=1)
data6
```

```python
#levels ohe
data7 = pd.concat([data6,pd.get_dummies(data2['levels'], prefix='levels')],axis=1)
data7
```

```python
#homeType ohe
data8 = pd.concat([data7,pd.get_dummies(data2['homeType'], prefix='homeType')],axis=1)
data8
```

```python
# county ohe
data9 = pd.concat([data8,pd.get_dummies(data2['county'], prefix='county')],axis=1)
data9
```

```python
# removing the original categorical columns
data11 = data9.drop(
    columns=[
        "state",
        "city",
        "event",
        "levels",
        "homeType",
        "lotAreaUnits",
        "county"
    ]
)
data11
```

```python
#subsetting for only ca data
ca_subset = data11[data11["state_CA"]==1].copy()
ca_subset
```

```python
#extracting the prices (our ys)
y_values_CA = ca_subset["price"]
y_values_CA
```

```python
# removing the y 
ca_subset = ca_subset.drop(
    columns=[
        "price"
    ]
)
ca_subset
```

```python
#subsetting for only ga data
ga_subset = data11[data11["state_GA"]==1].copy()
ga_subset
```

```python
#extracting the prices (our ys)
y_values_GA = ga_subset["price"]
y_values_GA
```

```python
# removing the y 
ga_subset = ga_subset.drop(
    columns=[
        "price"
    ]
)
ga_subset
```

```python
from sklearn.model_selection import train_test_split

X_train_CA, X_test_CA, y_train_CA, y_test_CA = train_test_split(ca_subset, y_values_CA, test_size=0.3)
X_train_GA, X_test_GA, y_train_GA, y_test_GA = train_test_split(ga_subset, y_values_GA, test_size=0.3)
```

```python
#combing to get data for both states
X_train_All = pd.concat([X_train_CA, X_train_GA])
X_test_All = pd.concat([X_test_CA, X_test_GA])
y_train_All = pd.concat([y_train_CA, y_train_GA])
y_test_All = pd.concat([y_test_CA, y_test_GA])
```

```python
# standardizing the numeric variables in the training data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_CA[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler.fit_transform(
    X_train_CA[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_train_CA
```

```python
# standarding the testing data from the training data
X_test_CA[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler.transform(
    X_test_CA[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_test_CA
```

```python
# standardizing the numeric variables in the training data
from sklearn.preprocessing import StandardScaler

scaler2= StandardScaler()

X_train_GA[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler2.fit_transform(
    X_train_GA[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_train_GA
```

```python
# standarding the testing data from the training data
X_test_GA[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler2.transform(
    X_test_GA[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_test_GA
```

```python
# standardizing the numeric variables in the training data
from sklearn.preprocessing import StandardScaler

scaler3= StandardScaler()

X_train_All[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler3.fit_transform(
    X_train_All[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_train_All
```

```python
# standarding the testing data from the training data
X_test_All[
    ["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]
] = scaler3.transform(
    X_test_All[["livingArea", "bathrooms", "bedrooms", "yearBuilt", "garageSpaces"]]
)
X_test_All
```

```python
#splitting the training data into further training and validation data for model comparison
X_train_model2, X_valid_model2, y_train_model2, y_valid_model2 = train_test_split(X_train_All, y_train_All, test_size=0.3)
```
