## Cleaning Data

```python
#Checking for any missing values
all_data.isnull().values.any()
```

```python
#checking which columns have NAs
all_data.isna().any()
```

```python
#checking which rows having NAs
all_data[all_data.isna().any(axis=1)]
```

```python
#subset to disclude events for rent
all_data2 = all_data[all_data["event"] != "Listed for rent"]
all_data2
```

```python
test = all_data2[all_data2["hasBadGeocode"] ==1]
test
```

```python
#removing lots from the data
all_data3 = all_data2[all_data2["homeType"] != "LOT"]
all_data3
```

```python
#removing observations with bad geocodes
all_data4 = all_data3[all_data3["hasBadGeocode"] == 0]
all_data4
```

```python
#checking for duplicates
duplicate = all_data4[all_data4.duplicated(subset=['id'])]
duplicate
```

I looked at many of the duplicated rows and they all had the same exact values for every feature, so I decieded to keep the last row for every duplicated observation.

```python
#dropping the duplicates
all_data5 = all_data4.drop_duplicates(
    subset=["id"], keep="last", inplace=False
).copy()
all_data5
```

```python
#checking for duplicates again
all_data5.duplicated(subset=['id']).any()
```

```python
#dropping unneeded columns
all_data6 = all_data5.drop(columns=
    [
        "Unnamed: 0",
        "id",
        "country",
        "datePostedString",
        "time",
        "pricePerSquareFoot",
        "cityId",
        "stateId",
        "streetAddress",
        "zipcode",
        "latitude",
        "longitude",
        "description",
        "currency",
        "hasBadGeocode",
        "countyId"

    ]
)

all_data6
```

```python
#Checking for any missing values in final dataset
all_data6.isnull().values.any()
```

```python
# finding relationship between price and variables
price_corr = all_data6.corr()["price"]
price_corr.abs().sort_values(ascending=False)[1:]
```

```python
all_data6.info()
```

```python
# transfrom non-int variables to int
all_data6[['bathrooms', 'bedrooms', 'garageSpaces']] = all_data6[['bathrooms', 'bedrooms', 'garageSpaces']].astype('int64')
```

```python
# drop the 'livingAreaValue' column, it's almost the same as 'livingArea'
all_data6 = all_data6.drop('livingAreaValue', axis=1)
```

```python
all_data6.is_bankOwned.value_counts()
```

```python
all_data6.is_forAuction.value_counts()
```

```python
# drop the 'is_bankOwned' column, there are only 4 bank owned houses out of 33743
#          'is_forAuction',there are only 18 houses for auction out of 33743
all_data6 = all_data6.drop(['is_bankOwned', 'is_forAuction'], axis=1)
```

```python
# drop the 'buildingArea' column, because
# 1. most of the building area is 0
print(all_data6.buildingArea[all_data6['buildingArea']==0].count(), '\n')

# 2. except for rows that are 0, the values of buildingArea and livingArea are exactly the same
temp = all_data6[['livingArea','buildingArea', 'lotAreaUnits']]

temp = temp[temp['buildingArea']!=0]
print(temp.livingArea.describe(), '\n')
print(temp.buildingArea.describe())

all_data6 = all_data6.drop('buildingArea', axis=1)
```

```python
# remove outliers - year
all_data6['yearBuilt'] = all_data6['yearBuilt'].replace([0, 9999], np.NaN) 
all_data6['yearBuilt'] = all_data6['yearBuilt'].fillna(all_data6['yearBuilt'].mean())
all_data6['yearBuilt'] = all_data6['yearBuilt'].astype('int64')
all_data6.shape
```

```python
# remove outliers - bathrooms
all_data6.bathrooms.unique()
```

```python
bath_out = (all_data6.bathrooms == 0) | (all_data6.bathrooms > 20)

all_data6 = all_data6[~bath_out]
all_data6.shape
```

```python
# remove outliers - bedrooms
all_data6.bedrooms.unique()
```

```python
bed_out = (all_data6.bedrooms == 0) | (all_data6.bedrooms > 20)

all_data6 = all_data6[~bed_out]
all_data6.shape
```

```python
all_data6.columns
```

```python
# minimize categories of 'levels'
all_data6.levels.unique()
```

```python
# re-categorize levels 
one_level = ['One Story', 'One', 'One Story-One', '1', 'Other-One', 'Manufactured Home 1 Story', 
            'One-Manufactured Home 1 Story', 'One-Mobile Home 1 Story', 'One-Two-Mobile Home 1 Story']
two_levels = ['Two Story', 'Two', 'Two Story-Two', 'Two Story-One', '2', 'One and One Half', 
            'One-Two Story Foyer', 'Two-Two Story Foyer', 'Two-Foyer - 2 Story']
split_level = ['Multi/Split', 'Tri-Level', 'Three Or More-Multi/Split', 'Two-Multi/Split', 
            'Multi/Split-Tri-Level', 'Tri-Level-Two', 'Multi/Split-Three Or More', 'One-Multi/Split', 
            'Multi/Split-One', 'Multi/Split-Two', 'Split Level', 'One-Two-Multi/Split', 'Three Or More-Split Level', 
            'Two-Three Or More-Multi/Split', 'Two-Split Foyer', 'Split Foyer-Split Level', 'Split Foyer', 
            'One and One Half-Split Level', 'Tri Level-Split Level', 'Two-Split Level', 'Split Level-Two', 
            'Three Or More-Split Level-Tri-Level', 'One and One Half-Multi/Split', 'One-Two', 'One Story-Two', 
            'Two-One', 'One and One Half-Two', 'One-One and One Half', 'One-One and One Half-Two', 'One-Other']
others = ['Three Or More', 'Three', '0', 'One Story-Three Or More', 'Four', 'Three or More Stories', 
            'Three or More Stories-Three Or More', 'Other', 'Three or More Stories-One', 'One-Two-Three Or More', 
            'Two-Three Or More', 'Two Story-Three Or More', 'Three or More Stories-Two', 'One-Three Or More', 
            '3', '4+', 'Tri-Level-Three Or More', 'Multi-Level', 'Three Or More-Two', 'Three or More Stories-One-Two', 
            'Multi/Split-Split Level', '2 Story or More', 'Tri Level', '2.5 Story', 'Over 2 Stories', 
            'Three Or More-Two Story Foyer', 'Other-See Remarks', '3 Story', 'Two and One-Half']

all_data6['levels'] = all_data6['levels'].replace(one_level, 'one level') 
all_data6['levels'] = all_data6['levels'].replace(two_levels, 'two levels') 
all_data6['levels'] = all_data6['levels'].replace(split_level, 'split level') 
all_data6['levels'] = all_data6['levels'].replace(others, 'others') 
```
